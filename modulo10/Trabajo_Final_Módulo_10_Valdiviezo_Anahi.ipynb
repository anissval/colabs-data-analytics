- **l2** - provincia o ciudad
- **l3** - barrio

# **Creación del DataFrame**

```
url="https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/"
archivo="propiedades.csv"
datos = pd.read_csv(url+archivo, index_col=0)
```

```
datos.dropna(how="any", inplace=True)
```

```
nombres_columnas = ["Barrio", "Ambientes", "Habitaciones", "Baños", "Sup_Total", "Sup_Cubierta", "Precio"]
```

```
datos.columns = nombres_columnas 
```

# **Generación del modelo de regresión**

**Configuración del entorno PyCaret**

```
regresion = setup(data = datos, 
                  target = "Precio",
                  train_size = 0.80, 
                  session_id=123)
```

```
regresion.dataset
```

```
models()
```

```
print(f"Los modelos de regresión son {len(models())}")
```

### **Comparando modelos**

```
modelos = compare_models()
```

### **Creacion del modelo**

```
modelo_01 = create_model('et')
```

### **Optimizando**

```
optimizar_modelo_01 = tune_model(modelo_01)
```

```
interpret_model(modelo_01)
```
